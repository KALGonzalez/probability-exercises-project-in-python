## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO

import random

# Número de simulaciones
num_simulaciones = 1000

# Contador de casos favorables
casos_favorables = 0

# Realiza las simulaciones
for _ in range(num_simulaciones):
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    suma = dado1 + dado2
    
    # Verifica si la suma es mayor a 7 o es un número par
    if suma > 7 or suma % 2 == 0:
        casos_favorables += 1

# Calcula la probabilidad estimada
probabilidad_estimada = casos_favorables / num_simulaciones

print(f"La probabilidad estimada de que la suma sea mayor a 7 o un número par es: {probabilidad_estimada:.4f}")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
```

```python
import random

# Definir el contenido de la caja
bolas = ['blanca'] * 10 + ['roja'] * 20 + ['verde'] * 30

# Número de simulaciones
num_simulaciones = 1000

# Contadores de casos favorables
casos_3_blancas_2_rojas = 0
casos_todas_mismo_color = 0

# Realiza las simulaciones
for _ in range(num_simulaciones):
    muestra = [random.choice(bolas) for _ in range(5)]
    
    # Contar las bolas de cada color en la muestra
    blancas = muestra.count('blanca')
    rojas = muestra.count('roja')
    verdes = muestra.count('verde')
    
    # Verificar la condición de 3 blancas y 2 rojas
    if blancas == 3 and rojas == 2:
        casos_3_blancas_2_rojas += 1
        
    # Verificar la condición de todas del mismo color
    if blancas == 5 or rojas == 5 or verdes == 5:
        casos_todas_mismo_color += 1

# Calcular las probabilidades estimadas
probabilidad_3_blancas_2_rojas = casos_3_blancas_2_rojas / num_simulaciones
probabilidad_todas_mismo_color = casos_todas_mismo_color / num_simulaciones

print(f"La probabilidad estimada de tomar 3 blancas y 2 rojas es: {probabilidad_3_blancas_2_rojas:.4f}")
print(f"La probabilidad estimada de tomar todas las bolas del mismo color es: {probabilidad_todas_mismo_color:.4f}")
```
